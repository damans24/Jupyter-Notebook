## Subplot

Subplots are groups of smaller axes that can exist together within a single figure for each comparison.

```python
# subplot(# of rows, # of cols, index): a 2x3 grid with 6 subplots is illustrated below:
Image('../img/subplot.png')
```

```python
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 4)
ax1.plot(x, x*2)
ax2.plot(x, x**2)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,15))
ax[0, 0].plot(x, x*2)
ax[1, 1].plot(x, x**2)
```

```python
#subplot(ax): ax specifies the number of subplots and return a list of axes
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# Plot the Normal Distribution
# you can include Latex inline by enclosing latex text with $, such as $\mu$
from scipy.stats import norm

fig, ax = plt.subplots(figsize=(10,10))

x = np.linspace(-10, 10, 1000)
mu = 0
sigma = 2
dist = norm(mu, sigma)
ax.plot(x, dist.pdf(x), label=f'$\mu={mu}, \sigma={sigma}$')

ax.set_title('Normal Distribution', fontsize=20)
ax.set_xlabel('$x$')
ax.set_ylabel('$p(x|\mu,\sigma)$')

ax.legend()
```

Create a plot that overlays four normal distributions: 
1. $\mu$=0, $\sigma$=1
2. $\mu$=0, $\sigma$=2
3. $\mu$=0, $\sigma$=3

```python
# Overlay multiple normal distributions in the same plot
# you can include Latex inline by enclosing latex text with $, such as $\mu$


# Add your code here: 
fig = plt.figure(figsize=(10,10))
ax = plt.axes()

x = np.linspace(-10, 10, 1000)

dist = norm(0,1)
ax.plot(x, dist.pdf(x), label = f'$/mu={0}, /sigma={1}$')

dist = norm(0,2)
ax.plot(x, dist.pdf(x), label = f'$/mu={0}, /sigma={2}$')

dist = norm(0,3)
ax.plot(x, dist.pdf(x), label = f'$/mu={0}, /sigma={3}$')

dist = norm(2,1)
ax.plot(x, dist.pdf(x), label = f'$/mu={2}, /sigma={1}$')

ax.set_xlabel('$x$')
ax.set_ylabel('$p(x|\mu,\sigma)$')
ax.set_title('Normal Distributions', fontsize=20)
```

## Figure Size, Labels, Legend

- Specify different figure size
- add labels for each plot
- add legend for the figure
- adjust the axes limits (the default ones are often good enough)

```python
Image('../img/matplotlib-figure-anatomy.png')
```

```python
# figsize=(width, height) specifies figure size in inches, default to (6.4, 4.8)
fig, ax = plt.subplots(figsize=(10,10))

x = np.linspace(0, 10, 50) # return 50 numbers between 0 and 10

# set the title and axes labels

ax.set_title('Figure for Demonstration', fontsize=20)
ax.set_xlabel('The value of x')
ax.set_ylabel('The value of y')

# set axes limits
ax.set_xlim(-3, 12)
ax.set_ylim(-200, 1200)

# set plot lable and show legend
ax.plot(x, x**3, label='y=x^3') 
ax.plot(x, x**3 - 8*x**2 + 5, label='y=x^3-8x^2+5') 
ax.plot(x, 2**x, label='y=2^x')

ax.legend()
```

## Line Style and Color
- Styles: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
- Colors: https://matplotlib.org/stable/tutorials/colors/colors.html

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes()
# line style
ax.plot(x, x, linestyle='solid')
ax.plot(x, x + 1, linestyle='dashed')
ax.plot(x, x + 2, linestyle='dashdot')
ax.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
ax.plot(x, x + 5, linestyle='-')  # solid
ax.plot(x, x + 6, linestyle='--') # dashed
ax.plot(x, x + 7, linestyle='-.') # dashdot
ax.plot(x, x + 8, linestyle=':');  # dotted

# color
ax.plot(x, x + 10, color='blue')        # specify color by name
ax.plot(x, x + 11, color='g')           # short color code (rgbcmyk)
ax.plot(x, x + 12, color='0.75')        # Grayscale between 0 and 1
ax.plot(x, x + 13, color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
ax.plot(x, x + 14, color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
ax.plot(x, x + 15, color='chartreuse'); # all HTML color names supported

# combine line style and color
ax.plot(x, x + 17, '-g')  # solid green
ax.plot(x, x + 18, '--c') # dashed cyan
ax.plot(x, x + 19, '-.k') # dashdot black
ax.plot(x, x + 20, ':r');  # dotted red
```

## Exercises: 

Create a figure that overlays four histograms that shows one of the following probability distributions: 
1. Normal distribution: $\mu$=0, $\sigma$=1
2. Normal distribution: $\mu$=0, $\sigma$=2
3. Uniform distrubtion between -2 and 2
4. Exponential distrubtion with rate parameter $\lambda=\frac{1}{\beta}$=2

Hint: 
- Uniform distrubiton: https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html
- Exponential distrubtion: https://numpy.org/doc/stable/reference/random/generated/numpy.random.exponential.html

```python
# Add your code here: 

normal1 = np.random.normal(loc=0, scale=1, size=1000)
normal2 = np.random.normal(loc=0, scale=2, size=1000)
uniform = np.random.uniform(low=-2, high=2, size=1000)
exponential = np.random.exponential(scale=1/2, size=1000)
fig, ax = plt.subplots(figsize=(10,6))
ax.hist(normal1, bins=30, alpha=0.5, label='N(0,1)')
ax.hist(normal2, bins=30, alpha=0.5, label='N(0,2)')
ax.hist(uniform, bins=30, alpha=0.5, label='U(-2,2)')
ax.hist(exponential, bins=30, alpha=0.5, label='Exp(1/2)')
ax.legend()
ax.set_title('Overlaying Histogram of Four Distributions')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
```
